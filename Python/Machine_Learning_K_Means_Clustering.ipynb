Using machine learning k-means clustering

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
data = (r"/Ch5_bike_station_locations.csv")
df = pd.read_csv(data, encoding='latin-1')
df.head()
```

```python
plt.scatter(df['latitude'], df['longitude'])
plt.show()
```

```python
from sklearn.cluster import KMeans
clusdata = list(zip(df['latitude'], df['longitude']))
inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(clusdata)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Based on this graph, 2 seems like it will be an appropriate amount of clusters.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(df[['latitude', 'longitude']])
df['cluster'] = kmeans.labels_
plt.scatter(df['latitude'], df['longitude'], c=df['cluster'])
plt.show()
```
